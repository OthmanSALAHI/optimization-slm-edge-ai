```python
import torch
from transformers import AutoTokenizer, T5ForConditionalGeneration

def launch_mini_chatbot():
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
    MODEL_PATH = "./outputs/mini_chatbot_model"
    BASE_MODEL = "google/flan-t5-small"

    print("🤖 Initializing Edge AI Mini-Chatbot engine...")
    tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
    
    # Attempt to load your custom fine-tuned weights; fallback to baseline if missing
    try:
        model = T5ForConditionalGeneration.from_pretrained(MODEL_PATH).to(DEVICE)
        print("🔥 Loaded your optimized fine-tuned weights successfully!")
    except Exception:
        print("⚠️ Fine-tuned folder not found. Falling back to baseline Flan-T5-Small.")
        model = T5ForConditionalGeneration.from_pretrained(BASE_MODEL).to(DEVICE)
        
    model.eval()  # Set to evaluation mode to lock down dropout layers
    
    print("\n💬 Chatbot initialized. Type 'exit' or 'quit' to close the session.\n" + "─"*60)
    
    while True:
        try:
            user_input = input("\nYou 👤: ")
            if user_input.lower() in ['exit', 'quit']:
                print("🤖 Chatbot: Terminating session. Goodbye!")
                break
                
            if not user_input.strip():
                continue
            
            # Formatting input clearly to prompt an instruction-following response
            prompt = f"Respond to the following short query: {user_input}"
            
            # Tokenize user text
            inputs = tokenizer(prompt, return_tensors="pt").to(DEVICE)
            
            # Generate the chatbot response token sequence
            with torch.no_grad():
                outputs = model.generate(
                    input_ids=inputs["input_ids"],
                    attention_mask=inputs["attention_mask"],
                    max_new_tokens=64,     # Keeps responses short and sweet for edge deployment
                    do_sample=True,        # Enables creative sampling over rigid deterministic text
                    top_p=0.92,            # Nucleus sampling to weed out low-probability gibberish
                    temperature=0.7        # Balances focus with casual conversational variation
                )
            
            # Decode back to readable text
            response = tokenizer.decode(outputs[0], skip_special_tokens=True)
            print(f"Chatbot 🤖: {response}")
            print("─"*60)
            
        except KeyboardInterrupt:
            print("\n🤖 Chatbot: Force closed. Goodbye!")
            break

if __name__ == "__main__":
    launch_mini_chatbot()
```
